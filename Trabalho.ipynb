```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plotar(valores, titulo, b=0):
    fig = plt.figure(figsize=(18, 1), dpi= 80, facecolor='w', edgecolor='k')
    plt.plot(valores, label="Valor Original")
    plt.title(titulo)
    if b!=0:
        plt.plot(b, label="Novo Valor")
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
def detalhes(valores):
    print("max: {}\nmin: {}\nvariaçao: {}".format(max(valores), min(valores), max(valores)-min(valores)))
```

```python
def comparar(valores, b):
    _ = ["J", "Q", "N"]
    for i in range(3):
        print(_[i])
        plotar(valores[i],_[i], b[i])
        detalhes(valores[i])
        detalhes(b[i])
```

```python
def mostrar(valores):
    _ = ["J", "Q", "N"]
    for i in range(3):
        print(_[i])
        plotar(valores[i], _[i])
        detalhes(valores[i])
```

```python
def simulacao(tn, q, k, k1, k2, k3):
    result_q = []
    result_n = []
    result_j = []
    for t in range(500):
        j = 40*math.sin(t/15.9)
        if j < 0:
            j = 0
        n = tn - k3*q
        jr = j/(1+k*n)
        dq = k1*jr*n-k2*q
        q += dq
        result_q.append(q)
        result_n.append(n)
        result_j.append(j)
    return [result_j, result_q, result_n]
```

```python
resultado_a = simulacao(tn=1.0, q=5.0, k=.9, k1=.2, k2=.04, k3=.01)
mostrar(resultado_a)
```

```python
resultado_b = simulacao(tn=0.3, q=5.0, k=.9, k1=.2, k2=.04, k3=.01)
comparar(resultado_a, resultado_b)
```

```python
resultado_c = simulacao(tn=1.3, q=5.0, k=.9, k1=.2, k2=.04, k3=.01)
comparar(resultado_a, resultado_c)
```

```python
resultado_d = simulacao(tn=1.0, q=5.0, k=.9, k1=.2, k2=.08, k3=.01)
comparar(resultado_a, resultado_d)
```
